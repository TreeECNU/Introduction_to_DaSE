```python
#判断输入a是否是质数
import math
def whether_prime(num):
    index = 0
    if num == 2:
        index = 1
        return index
    for i in range(2, int(math.sqrt(num) + 1), 1):
        if num % i == 0:
            index = 1
            break
    return index
num = int(input("input = "))
res = whether_prime(num)
if res == 1:
    print("It is a prime.")
else:
    print("It isn't a prime.")
```

```python
#插入排序
def insertion_sort(nums):
    for i in range(1, len(nums), 1):
        index = nums[i]
        for j in range(i - 1, -1, -1):
            if nums[j] > index:
                nums[j + 1] = nums[j]
            if nums[j] <= index:
                break
        nums[j + 1] = index

nums = [1, 7, 2, 2, 8, 3]
insertion_sort(nums)
print(nums)
```

```python
#希尔排序
def insertion_sort(nums, a, b):
    for i in range(a + 1, b + 1, 1):
        index = nums[i]
        for j in range(i - 1, a - 1, -1):
            if nums[j] > index:
                nums[j + 1] = nums[j]
            if nums[j] <= index:
                break
        nums[j + 1] = index
def shell_sort(nums):
    n = len(nums) // 2
    while n > 0:
        i = 0
        while i <= len(nums)- n:
            insertion_sort(nums, i, i + n - 1)
            i += 1
        n = n // 2

nums = [1, 3, 2, 9, 4, 7, 5, 6]
shell_sort(nums)
print(nums)
```

希尔排序的平均时间复杂度为：$O(n*log^2n)$  
希尔排序的最坏时间复杂度为：$O(n^2)$  
希尔排序的空间复杂度为：$O(1)$
