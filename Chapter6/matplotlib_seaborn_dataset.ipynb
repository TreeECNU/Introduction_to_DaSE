```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib as mpl
from scipy.stats import norm
import seaborn as sns
import pandas as pd
sns.set(style="darkgrid") #这是seaborn默认的风格

# 中文乱码的处理
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']# 设置微软雅黑字体
plt.rcParams['axes.unicode_minus'] = False # 避免坐标轴不能正常的显示负号
```

```python
# 读入数据
iris = pd.read_csv(r'iris.csv')
# 绘制散点图
plt.scatter(x = iris.Petal_Width, # 指定散点图的x轴数据
            y = iris.Petal_Length, # 指定散点图的y轴数据
            color = 'blue' # 指定散点图中点的颜色
           )
# 添加x轴和y轴标签
plt.xlabel('花瓣宽度')
plt.ylabel('花瓣长度')
# 添加标题
plt.title('鸢尾花的花瓣宽度与长度关系')
# 显示图形
plt.show()
```

```python
# seaborn模块绘制分组散点图
sns.lmplot(x = 'Petal_Width', # 指定x轴变量
           y = 'Petal_Length', # 指定y轴变量
           hue = 'Species', # 指定分组变量
           data = iris, # 指定绘图数据集
           legend_out = False, # 将图例呈现在图框内
           truncate=True # 根据实际的数据范围，对拟合线作截断操作
          )
# 修改x轴和y轴标签
plt.xlabel('花瓣宽度')
plt.ylabel('花瓣长度')
# 添加标题
plt.title('鸢尾花的花瓣宽度与长度关系')
# 显示图形
plt.show()
```

```python
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
```

```python
# matplotlib 画图
plt.plot(x, y, linestyle='--', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Seaborn画图
df = pd.DataFrame({'x':x, 'y':y})
sns.lineplot(x="x", y="y", data=df, linestyle = "--", marker="o")
plt.show()
```

```python
a = np.random.randn(100)
s = pd.Series(a)
```

```python
# matplotlib绘图
plt.hist(s)
plt.show()
```

```python
# Seaborn绘图
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
# kde是核密度函数估计
```

```python
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]
```

```python
# matplotlib绘制图
plt.bar(x, y, align='center', color='steelblue', alpha=0.85)
plt.show()
```

```python
# Seaborn绘制图
data = pd.DataFrame({"labels":x, "labelsnum":y})
sns.barplot(x = "labels", y ="labelsnum", data=data)
plt.show()
```

```python
# 数据准备
# 生成0-1之间的10*4维度数据
data = np.random.normal(size=(10, 4))
labels = ['A', 'B', 'C', 'D']
```

```python
# matplotlib绘制箱线图
plt.boxplot(data, labels=labels)
plt.show()
```

```python
# Seaborn绘制箱线图
df = pd.DataFrame(data, columns=labels)
sns.boxplot(data=df)
```
